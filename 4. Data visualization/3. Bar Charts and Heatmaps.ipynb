# 데이터 불러오기

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
'''
Month라는 index가 Datetime 타입이 아니기에, parse_dates=True를 지정할 수 없음.
'''

flight_filepath = "./data/flight_delays.csv"
flight_data = pd.read_csv(flight_filepath, index_col="Month")
flight_data.head()
```

```python
'''
궁금: 여기서 3월의 최소 비행기에 대한 코드를 간단히 짜는 방법은 뭘까?
'''

flight_data.loc[3][flight_data.loc[3] == min(flight_data.loc[3])].index.values[0]
```

# Bar chart

```python
plt.figure(figsize=(10,6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# barplot 그리기
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# y라벨 지정
plt.ylabel("Arrival delay (in minutes)")
```

# Heatmap
 - annot=True: Heatmap위에 수치를 명시
 - 색깔로 눈에 띄는 구간을 쉽게 식별 가능(9-11월)

```python
plt.figure(figsize=(14,7))
plt.title("Average Arrival Delay for Each Airline, by Month")
plt.xlabel("Airline")

# Heatmap 그리기
sns.heatmap(data=flight_data, annot=True)
```
