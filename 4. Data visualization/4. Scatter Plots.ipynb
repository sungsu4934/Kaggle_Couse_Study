# Load data

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
insurance_filepath = "./data/insurance.csv"
insurance_data = pd.read_csv(insurance_filepath)
insurance_data
```

# scatterplot & regplot
 - 상관관계 확인 가능
 - 더 자세한 결과를 알고싶다면, regression line까지 함께 plot해보자.
 - sns.scatterplot: 점
 - sns.regplot: 점 + 선
 
##### 주의: 데이터의 개수가 적다면, 지나치게 그래프를 믿지 마라!

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

##### 색상까지 추가하여, 기존 BMI, charges 이외에 1가지 인자(smoker)를 더 고려하여 시각화 가능

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

# lmplot
#### : 이전에는 전체 집단에 대한 1개의 regplot을 그렸다면,  lmplot은 색으로 구분한 각 집단에 대해 각각 regplot을 형성
 - 코드 구성시, data 파라미터를 지정하여, x, y에 반복되는 값을 피할 수 있다.

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

# swarmplot
#### : categorical scatter plot
 - 카테고리형 데이터에는 산점도가 그려질 수 없음.
 - 이를 보완한 Graph

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```
