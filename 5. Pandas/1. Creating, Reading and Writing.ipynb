# Creating data

```python
import pandas as pd
'''
최대 줄수 및 열수 설정 꿀팁!
'''
pd.set_option('display.max_rows', 10) # 최대 줄 수 설정
pd.set_option('display.max_columns', 10) # 최대 열 수 설정
```

##### Dictionary의 key가 column명으로, value는 값으로!

```python
'''
만약 행이 1개인 df를 만들더라도, list형태로 value를 구성해주자.
'''
pd.DataFrame({'Yes':[50, 21], 'No':[131, 2]})
```

##### dtype은 int외에 string도 가능하다.

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']})
```

##### default값인 0, 1... 등의 index 이외에 내가 하고싶은 index를 설정

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

### Series
 - Dataframe의 한 열
 - 여러 시리즈가 모이면 Dataframe이 된다!

```python
pd.Series([1, 2, 3, 4, 5])
```

##### Series의 index명도 내 마음대로 변경 가능!
 - Series에는 column (x) --> name (o)

```python
pd.Series([30, 35, 40], index=['A','B','C'], name='Product A') # name이 concat시 df에서 열이름으로 들어간다.2
```
