# 결측치 핸들링
 - 어떤 열에 대해 결측치가 있는 행 boolean indexing: pd.isnull()
 - 결측치 대체: fillna()
 - 결측치가 있는 상태에서는 value_counts() 함수가 오류 발생하기에 대체를 해준 후 활용하자.

```python
# pd.isnull()을 활용한 boolean indexing: 결측치가 있는 행 불러오기
reviews[pd.isnull(reviews.country)]
reviews[reviews.country.isnull()]
reviews[reviews.country.notnull()]
```

```python
# country 열에서 결측치의 개수는?
print('country열의 결측치 개수: '+ str(reviews[pd.isnull(reviews.country)].shape[0]))
```

```python
# region_2 열의 결측치를 Unknown으로 대체!
reviews.region_2.fillna("Unknown")
```

```python
# 결측치 대체 후 value_count

reviews.region_2.fillna("Unknown").value_counts()
```
