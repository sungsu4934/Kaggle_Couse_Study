# 기본세팅

```python
import pandas as pd
reviews = pd.read_csv('data/winemag-data-130k-v2.csv', index_col=0)
reviews
```

# groupby

```python
# points로 데이터를 그룹화 한 후, 각 개수를 파악!
reviews.groupby('points').points.count()
```

```python
# points로 그룹화 한 후, price column에 대한 최소값을 파악
reviews.groupby('points').price.min()
```

```python
# groupby에 lambda와 apply를 활용하여 함수까지 활용가능
'''
이거 원리가 뭘까,,,,,,,,
'''

reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
'''
이것도 원리가 와닿지 않음 --> groupby 후, point가 max인 것만 불러온 것 같기한데,,,
'''
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
# agg를 활용해 여러 값을 집계 가능
reviews.groupby(['country']).price.agg([len, min, max]) # len은 count의 의미
```

# Multi-index

```python
# groupby를 두 개로 하면 multiindex가 생긴다.
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed

'''
동일한 방법!
집단별 크기 (count)를 파악할때는 size함수가 good 
reviews.groupby(['country', 'province']).description.size()
'''
```

```python
# 묶은 후 값들에 대한 통계량 볼때 describe()함수는 용이하다!
countries_reviewed.describe()
```

```python
# multiindex 타입 확인!
mi = countries_reviewed.index
type(mi)
print(mi)
```

```python
# reset_index 사용시 두 개의 열이 생기면서 index가 초기화
countries_reviewed.reset_index()
```

# Sorting

```python
# len을 기준으로 sorting (by인자를 활용하며, default는 오름차순)
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
# 내림차순 적용
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
# index를 기준으로 정렬이 된다. -> sort_index
countries_reviewed.sort_index()
```

```python
# 기준 2가지로 sorting --> country로 우선 정렬하며, 동일한 country일 경우, len을 기준으로 정렬!
countries_reviewed.sort_values(by=['country', 'len'])
```
