# Mapping
 - apply와 map의 차이는 각 객체를 df에 넣느냐, series에 넣느냐 차이!!

### map 함수
 - 어떤 열에 대해 값을 대치할 때 for문을 많이 썼는데 시간 단축이 많이 될 것 같다.

```python
review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
```

```python
'''
궁금: map함수는 object type은 적용이 안되는걸까?
만약, 내가 taster_name열의 split 앞 단어를 가진 열을 만들고 싶다면 어떻게 할까?
'''
reviews.taster_name.map(lambda x: x.split(' ')[0])
```

### apply 함수
 - 함수를 정의한 후 df 전체에 적용이 가능

```python
'''
궁금: apply함수는 object type은 적용이 안되는걸까?
만약, 내가 taster_name열의 split 앞 단어를 가진 열을 만들고 싶다면 어떻게 할까? 그냥 for문이 답인가
'''

def get_first(row):
    row.taster_name = row.taster_name.split(' ')[0]
    return row

reviews.apply(get_first, axis='columns')
```

```python
# point 열에만 정의되도록 함수를 정의한 후, df에 전체 적용하면 알아서 된다.

def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns') # axis가 columns니까 columns에 대해 적용되는 것! --> index면 index명에 관련된 함수가 안에 들어가야함.
```

### 기타
 - .idxmax(): 최대인 index가 복수여도 1개만 반환하는 것이 단점일수도?

```python
# reviews.points / reviews.price 의 값이 최대인 index를 반환
(reviews.points / reviews.price).idxmax()
```

```python
# 각 열에서 어떤 문자열이 몇번 나왔는지 개수 산정
n_trop = reviews.description.map(lambda desc: "tropical" in desc).sum()
n_fruity = reviews.description.map(lambda desc: "fruity" in desc).sum()
descriptor_counts = pd.Series([n_trop, n_fruity], index=['tropical', 'fruity'])
descriptor_counts
```

```python
# apply 함수를 활용한 값 대치
def stars(row):
    if row.country == 'Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1

reviews.apply(stars, axis='columns')
```
