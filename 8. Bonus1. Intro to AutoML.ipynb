```python
PROJECT_ID = 'kaggle-playground-170215' # Project 명
BUCKET_NAME = 'automl-tutorial-alexis' # 구글 스토리지 버켓 이름

DATASET_DISPLAY_NAME = 'taxi_fare_dataset' # 데이터셋의 이름
TRAIN_FILEPATH = "../working/train_small.csv" # train data의 path
TEST_FILEPATH = "../input/new-york-city-taxi-fare-prediction/test.csv" # test data의 path

TARGET_COLUMN = 'fare_amount' # 타겟명
ID_COLUMN = 'key' # 구분코드

MODEL_DISPLAY_NAME = 'tutorial_model' # 모델명
TRAIN_BUDGET = 4000 # 1000: 1시간, 2000: 2시간... 얼마나 모델을 훈련하고싶은가?

# Import the class defining the wrapper
from automl_tables_wrapper import AutoMLTablesWrapper

# Create an instance of the wrapper
amw = AutoMLTablesWrapper(project_id=PROJECT_ID,
                          bucket_name=BUCKET_NAME,
                          dataset_display_name=DATASET_DISPLAY_NAME,
                          train_filepath=TRAIN_FILEPATH,
                          test_filepath=TEST_FILEPATH,
                          target_column=TARGET_COLUMN,
                          id_column=ID_COLUMN,
                          model_display_name=MODEL_DISPLAY_NAME,
                          train_budget=TRAIN_BUDGET)
```

```python
# Do not change: Create and train the model
amw.train_model()

# Do not change: Get predictions
amw.get_predictions()
```
